**LOGISTIC REGRESSION**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data =pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
legit=credit_card_data[credit_card_data.Class ==0]
fraud=credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=1000)
```

```python
new_dataset =pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5, stratify=y,random_state=99)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model = LogisticRegression()
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score on Training data :',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score on Test data :',test_data_accuracy)
```

**KNN ALGORITHM**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
credit_card_data =pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=500)
```

```python
new_dataset =pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
x = new_dataset.iloc[:,[2, 3]].values
y = new_dataset.iloc[:, -1].values
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, stratify=y,random_state=9)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)
```

```python
print('Confusion matrix :',cm)
```

```python
print('Accuracy score on Test data :',ac)
```
